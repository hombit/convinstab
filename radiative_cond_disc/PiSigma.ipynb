```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import vertstr

matplotlib.rc('text', usetex=True)


psis = [2.5, 3.5]
taus = np.logspace(0, 6, 12)

plt.xscale('log')
plt.xlabel(r'\tau_0')
plt.ylabel(r'\Pi_\Sigma')

for psi in psis:
    tau_plot = []
    PiSigma = []
    for tau in taus:
        try:
            Pi = vertstr.FindPi(tau, opacity=(1., psi)).getPi()
            tau_plot.append(tau)
            if psi == 2.5:
                PiSigma.append( Pi[0]**(1./18) * Pi[1]**(1./9) * Pi[2]**(7./9) * Pi[3]**(1./9) )
            elif psi == 3.5:
                PiSigma.append( Pi[0]**(1./20) * Pi[1]**(1./10) * Pi[2]**(4./5) * Pi[3]**(1./10) )
        except RuntimeError:
            pass
    if psi == 2.5:
        label = 'OPAL'
    elif psi == 3.5:
        label = "Kramer's"
    plt.plot( tau_plot, PiSigma, label=label )

plt.legend(loc='best')
```
