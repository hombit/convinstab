```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import vertstr
```

```python
tau = 1e4
fp = vertstr.FindPi(
    tau,
    heating='alpha',
    transfer='scattering'
)
sigma, pzqt = fp.pzqt()
delta_z = (1. - pzqt[:,3].min())**(-0.5) - 1
z = np.linspace(0, 1, 100)
theta_poly = (1-z**2/(1+delta_z)**2)
plt.xlim([0,1])
plt.ylim([0,1])
plt.title('log10(tau)={}'.format(np.log10(tau)))
plt.xlabel('z')
plt.ylabel('t')
plt.plot(pzqt[:,1], pzqt[:,3], label='Ketsaris-Shakura')
plt.plot(z, theta_poly, label='Polytrope')
plt.plot(z, 1-z**2, label='Polytrope')
plt.legend()
```
