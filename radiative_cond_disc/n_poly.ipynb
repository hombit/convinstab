```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import vertstr
from scipy import integrate, optimize
```

```python
tau = np.logspace(4, 12, 17)
n1 = np.empty_like(tau)
for i, t in enumerate(tau):
    fp = vertstr.FindPi(
        t,
        heating='alpha',
        transfer='scattering'
    )
    Pi = fp.getPi()
    n1[i] = Pi[0] / 2 - 1
    print('tau = ', np.log10(t), 'n1 = ', n1[i], 'Pi1..4 = ', Pi)

plt.xscale('log')   
plt.plot(tau, n1)
```
